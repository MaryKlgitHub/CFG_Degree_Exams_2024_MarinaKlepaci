```python
# 1 (2 marks). Load the netflix_titles.csv dataset as a Pandas dataframe:

import pandas as pd

# Load data
df = pd.read_csv('netflix_titles.csv')
```

```python
# Inspecting data
print(df.head())
print(df.info())
print(df.describe())
```

```python
# 2 (2 marks). Show the first 10 rows of the dataset
df.head(10)
```

```python
# 3 (1 mark). Show the last 5 rows
df.tail(5)
```

```python
# 4 (4 marks). Count the number of rows which have no missing value in any of the columns, and show the resulting data frame.

# Drop rows that contain any missing values
df_no_missing = df.dropna()

# Count the number of rows in this DataFrame
num_rows_no_missing = df_no_missing.shape[0]

# Output the count
print(f"Number of rows with no missing values: {num_rows_no_missing}")
print(df_no_missing) # it also shows the number of rows, just to shows other option to calculate
```

```python
# 5 (3 marks). Drop everything that is not a 'Movie' and show the resulting data frame. Use any method for this.

df_movies = df[df['type'] == 'Movie']
df_movies
```

```python
# Preparing to the next task

# Check the data type of the 'duration' column
print("Data type of 'duration':", df_movies['duration'].dtype)

# Check the first 5 rows of the 'duration' column
print("\nFirst 5 rows of 'duration':")
print(df_movies['duration'].head())

# Check the last 5 rows of the 'duration' column
print("\nLast 5 rows of 'duration':")
print(df_movies['duration'].tail())

# Check for missing values in the 'duration' column
missing_values_count = df_movies['duration'].isna().sum()
print(f"\nNumber of missing values in 'duration': {missing_values_count}")

```

```python
# 6 (7 marks). For every movie where the duration is missing, replace it with the average
# duration of the movies in the dataset and return the resulting data frame

# Convert the 'duration' column to string to handle mixed types
df_movies.loc[:, 'duration'] = df_movies['duration'].astype(str)

# Remove ' min' from the 'duration' column
df_movies.loc[:, 'duration'] = df_movies['duration'].str.replace(' min', '', regex=False)

# Convert the cleaned values to float
df_movies.loc[:, 'duration'] = pd.to_numeric(df_movies['duration'], errors='coerce')

# Replace missing values with the average duration
average_duration = df_movies['duration'].mean()
df_movies.loc[:, 'duration'] = df_movies['duration'].fillna(average_duration)

# Display the updated DataFrame
df_movies

# no calculating average cause no missing values in duration column

```

```python
# 7 (3 marks). Are there any duplicates in the data? If yes, please remove them.
duplicates = df_movies[df_movies.duplicated()]
num_duplicates = duplicates.shape[0]
num_duplicates

# No duplicates
```

```python
# 8 (2 marks). Rename all columns to start with a capital letter.
df_movies.columns = df_movies.columns.str.capitalize()
df_movies
```

```python
# 9 (1 mark). Save the cleaned dataset as a new csv file.
# Save the cleaned dataset as a new CSV file
df_movies.to_csv('cleaned_movies_dataset.csv', index=False)
```

2. MATPLOTLIB (25 points)

```python
# 1 (10 marks). Create a visualisation (by using the plot that is most appropriate for this) which shows the frequency of movies by year of release.

import matplotlib.pyplot as plt

# Count the number of movie releases per year
movies_by_year = df_movies['Release_year'].value_counts().sort_index()

# Create a bar plot to show frequency of movies by year
plt.figure(figsize=(14, 8))  # Set figure size for better clarity
plt.bar(movies_by_year.index, movies_by_year.values, color='skyblue')

# Adding title and labels
plt.title('Number of Movies Released Each Year')
plt.xlabel('Year of Release')
plt.ylabel('Number of Movies Released')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Add gridlines to improve readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
# 2 (15 marks). Create a pie chart which shows the 5 top most popular genres of the movis
  # in the dataset.

import pandas as pd
import matplotlib.pyplot as plt

# Extract and count genres
all_genres = df_movies['Listed_in'].str.split(',').explode().str.strip()
genre_counts = all_genres.value_counts()

# Identify the top 5 most popular genres
top_5_genres = genre_counts.head(5)

# Create a pie chart
plt.figure(figsize=(10, 7))
plt.pie(top_5_genres, labels=top_5_genres.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Top 5 Most Popular Genres')
plt.show()

```

```python
# 1 (2 marks). Running the code in this cell throws an error. Why do we get this error
  # and how do we fix it?

# use this numpy array for all exercises in this section
import numpy as np # need to define numpy lib
    
# use this numpy array for all exercises in this section
arr = np.array([0, 56, 82, 500, 140, 43, 92, 10, 25, 11, 18, 26, 150, 240, 364, 196])
arr
```

```python
# 2 (5 marks). How many odd numbers do we have in the array? Show how we get this result using numpy
odd = arr[arr % 2 != 0]
print(odd.size)
```

```python
# 3 (5 marks). Replace all odd numbers in arr with 999

arr[arr % 2 != 0] = 999

print("Updated array:", arr)
```

```python
# 4 (8 marks). The above is a 1D array. Convert it into a 4D array and reverse its columns.

# Desired output
# > array([[560,  82,  56,   0],
# >        [ 10,  92, 999, 140],
# >        [ 26,  18, 999, 999],
# >        [196, 364, 240, 150]])

# Reshape the array into a 4x4 2D array
array_4d = arr.reshape(4, 4)

print("Original 4D array:")
print(array_4d)

# Reverse the columns of the array
# `array_4d[:, ::-1]` means:
# - `:` selects all rows
# - `::-1` reverses the order of columns in each row
array_4d_reversed = array_4d[:, ::-1]

print("\n4D array with reversed columns:")
print(array_4d_reversed)
```
